```python
import pandas as pd
import plotly.graph_objects as go

# Load the dataset
data = pd.read_csv("/content/destination_folder/cleaned_dataset/metadata.csv")

# Verify dataset columns
print("Dataset columns:", data.columns)

# Ensure the required columns are present
required_columns = ['Capacity', 'Re', 'Rct']
missing_columns = [col for col in required_columns if col not in data.columns]
if missing_columns:
    raise ValueError(f"Missing columns: {missing_columns}")

# Extract relevant columns
capacity = data['Capacity']  # Capacity as X-axis
re = data['Re']              # Electrolyte Resistance as Y-axis
rct = data['Rct']            # Charge Transfer Resistance as Y-axis

# Create a colorful plot for Re (Electrolyte Resistance)
fig_re = go.Figure()
fig_re.add_trace(go.Scatter(
    x=capacity, y=re,
    mode='lines+markers',
    line=dict(color='rgb(26, 118, 255)', width=4, dash='dot'),  # Blue dashed line
    marker=dict(size=8, color='rgba(26, 118, 255, 0.7)', symbol='circle'),  # Blue transparent markers
    name="Electrolyte Resistance (Re)"
))
fig_re.update_layout(
    title={
        'text': "Electrolyte Resistance (Re) vs. Capacity",
        'y': 0.9,
        'x': 0.5,
        'xanchor': 'center',
        'yanchor': 'top'
    },
    xaxis=dict(title="Capacity", titlefont=dict(size=16, color="rgb(58, 71, 80)")),
    yaxis=dict(title="Re (Ohms)", titlefont=dict(size=16, color="rgb(58, 71, 80)")),
    template="plotly",
    font=dict(size=14, color="rgb(58, 71, 80)")
)

# Create a colorful plot for Rct (Charge Transfer Resistance)
fig_rct = go.Figure()
fig_rct.add_trace(go.Scatter(
    x=capacity, y=rct,
    mode='lines+markers',
    line=dict(color='rgb(255, 64, 64)', width=4, dash='dashdot'),  # Red dashdot line
    marker=dict(size=8, color='rgba(255, 64, 64, 0.7)', symbol='diamond'),  # Red transparent diamond markers
    name="Charge Transfer Resistance (Rct)"
))
fig_rct.update_layout(
    title={
        'text': "Charge Transfer Resistance (Rct) vs. Capacity",
        'y': 0.9,
        'x': 0.5,
        'xanchor': 'center',
        'yanchor': 'top'
    },
    xaxis=dict(title="Capacity", titlefont=dict(size=16, color="rgb(58, 71, 80)")),
    yaxis=dict(title="Rct (Ohms)", titlefont=dict(size=16, color="rgb(58, 71, 80)")),
    template="plotly",
    font=dict(size=14, color="rgb(58, 71, 80)")
)

# Display the plots
fig_re.show()
fig_rct.show()
```

```python
import pandas as pd
import plotly.graph_objects as go

# Load the dataset
data = pd.read_csv("/content/destination_folder/cleaned_dataset/metadata.csv")

# Verify the dataset columns
print("Dataset columns:", data.columns)

# Ensure the required columns are present
required_columns = ['Capacity', 'Re', 'Rct']
missing_columns = [col for col in required_columns if col not in data.columns]
if missing_columns:
    raise ValueError(f"Missing columns: {missing_columns}")

# Clean the 'Re' column by converting to numeric (any errors will be set to NaN)
data['Re'] = pd.to_numeric(data['Re'], errors='coerce')

# Handle NaN values in 'Re' column, for example, drop rows with NaN 'Re'
data = data.dropna(subset=['Re'])

# Create a Pie Chart for the 'Re' categories
# Let's assume we are categorizing 'Re' into ranges for the Pie chart
re_categories = pd.cut(data['Re'], bins=5, labels=["Low", "Medium Low", "Medium", "Medium High", "High"])

# Calculate the count for each category of 'Re'
re_category_counts = re_categories.value_counts()

# Create the Pie Chart
fig_pie = go.Figure(go.Pie(
    labels=re_category_counts.index,
    values=re_category_counts.values,
    hole=0.3,  # Donut chart effect
    textinfo='label+percent',  # Show label and percentage
    marker=dict(colors=['#FF9999', '#66B3FF', '#99FF99', '#FFCC99', '#FFB3E6'])
))

fig_pie.update_layout(
    title="Distribution of Electrolyte Resistance (Re) Categories",
    template="plotly"
)

# Show the Pie Chart
fig_pie.show()
```

```python
#########    1.Heatmap of Resistance Values Over Time ###########

import plotly.express as px
import pandas as pd

# Load the data
df = pd.read_csv('/content/destination_folder/cleaned_dataset/metadata.csv')

# Convert 'start_time' to datetime (specify the format to avoid UserWarning)
df['start_time'] = pd.to_datetime(df['start_time'], format='%Y-%m-%d %H:%M:%S', errors='coerce')  # Coerce invalid formats to NaT

# Sort the data by 'start_time'
df = df.sort_values('start_time')

# Create the 'battery_impedance' column as the sum of 'Re' and 'Rct'
df['battery_impedance'] = df['Re'] + df['Rct']

# Create a heatmap showing how Re, Rct, and battery_impedance change over time
df_heatmap = df[['start_time', 'battery_id', 'Re', 'Rct', 'battery_impedance']].copy()

# Convert 'start_time' to a string for better labeling
df_heatmap['start_time_str'] = df_heatmap['start_time'].dt.strftime('%Y-%m-%d %H:%M:%S')

# Melt the dataframe to make 'Re', 'Rct', and 'battery_impedance' values into a single column
df_melted = df_heatmap.melt(id_vars=['start_time_str', 'battery_id'],
                            value_vars=['Re', 'Rct', 'battery_impedance'],
                            var_name='parameter', value_name='value')

# Aggregate duplicate values by taking the mean (or use drop_duplicates as an alternative)
df_melted_agg = df_melted.groupby(['battery_id', 'start_time_str', 'parameter'], as_index=False).agg({'value': 'mean'})

# Create the heatmap using pivot with keyword arguments
df_pivoted = df_melted_agg.pivot(index='battery_id', columns='start_time_str', values='value')

# Create the heatmap
fig = px.imshow(df_pivoted,
                color_continuous_scale='Viridis', title="Heatmap of Battery Parameters Over Time")

fig.update_layout(xaxis_title='Time', yaxis_title='Battery ID', coloraxis_colorbar_title='Resistance (Ohms)')
fig.show()
```

```python
#######   4. Line Plot of Average Resistance per Cycle  ##########

import pandas as pd
import plotly.express as px

# Load the data
df = pd.read_csv('/content/destination_folder/cleaned_dataset/metadata.csv')

# Convert 'Re' and 'Rct' to numeric (this will convert non-numeric values to NaN)
df['Re'] = pd.to_numeric(df['Re'], errors='coerce')
df['Rct'] = pd.to_numeric(df['Rct'], errors='coerce')

# Create 'battery_impedance' as the sum of 'Re' and 'Rct'
df['battery_impedance'] = df['Re'] + df['Rct']

# Compute average values of resistance per cycle (test_id)
df_avg = df.groupby('test_id').agg({'Re': 'mean', 'Rct': 'mean', 'battery_impedance': 'mean'}).reset_index()

# Plot the average values for Re, Rct, and battery_impedance
fig = px.line(df_avg, x='test_id', y=['Re', 'Rct', 'battery_impedance'],
              title="Average Battery Parameters per Cycle",
              labels={'test_id': 'Cycle', 'value': 'Resistance (Ohms)'})
fig.show()
```

```python
#######   5.Subplots for Each Parameter  ##########


import plotly.subplots as sp
import plotly.graph_objs as go

# Create subplots for Re, Rct, and battery_impedance
fig = sp.make_subplots(rows=3, cols=1, shared_xaxes=True,
                       subplot_titles=('Electrolyte Resistance (Re)', 'Charge Transfer Resistance (Rct)', 'Battery Impedance'),
                       vertical_spacing=0.1)

# Add traces for Re, Rct, and battery_impedance
fig.add_trace(go.Scatter(x=df['test_id'], y=df['Re'], mode='lines', name='Re'), row=1, col=1)
fig.add_trace(go.Scatter(x=df['test_id'], y=df['Rct'], mode='lines', name='Rct'), row=2, col=1)
fig.add_trace(go.Scatter(x=df['test_id'], y=df['battery_impedance'], mode='lines', name='Battery Impedance'), row=3, col=1)

fig.update_layout(title="Battery Parameters Over Cycles", showlegend=False)
fig.show()
```
